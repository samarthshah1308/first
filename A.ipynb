```python
import random
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 
          'Nov', 'Dec']

total_sales = [random.randint(1000, 5000) for _ in range(12)]

plt.plot(months, total_sales, marker='o', linestyle='-', color='b', label='Total Sales')

plt.title('Monthly Sales Report')
plt.xlabel('Months')
plt.ylabel('Total Sales')
plt.grid(True)
plt.legend()

plt.show()
```
